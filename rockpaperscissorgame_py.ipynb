```python
import random
l=['rock','paper','scissor']
'''
rock vs paper
rock vs scissor
paper vs scissor
'''
while True:
  ccount=0
  ucount=0

  uc=int(input('''
       1 game start......
       2 end | exit
    '''))
  if uc==1:
      for a in range(1,6):
        userinput=int(input('''
1 rock
2 paper
3 scissor
              '''))
        if userinput==1:
            uchoice='rock'
        elif userinput==2:
            uchoice='paper'
        elif userinput==3:
            uchoice='scissor'
        Cchoice=random.choice(l)
        print(uchoice)
        print(Cchoice)
        if Cchoice==uchoice:
            print('computer value',Cchoice)
            print('user value',uchoice)
            print('draw')
            ucount=ucount+1
            ccount=ccount+1
        elif (uchoice=='rock'and Cchoice=='scissor')or(uchoice=='paper'and Cchoice=='rock')or(uchoice=='scissor'and Cchoice=='paper'):
            print('computer value',Cchoice)
            print('user value',uchoice)
            print('yu win')
            ucount=ucount+1

        else:

            print('computer value',Cchoice)
            print('user value',uchoice)
            print('computer wins')
            ccount=ccount+1

  if ucount==ccount:
        print('final game draw....')
        print('computer score',ccount)
        print('user score',ucount)
  elif ucount>ccount:
        print('final you win the game....')
        print('computer score',ccount)
        print('user score',ucount)
  else:
        print('final computer win the game....')
        print('computer score',ccount)
        print('user score',ucount)




else:
     print('thank you')

```
